```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/netflix_titles.csv")

df.head()
```

```python
df["type"].value_counts().plot(kind="bar")

plt.title("Movies vs TV Shows")
plt.savefig("../images/movies_vs_tvshows.png",bbox_inches="tight")
plt.show()
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries")
plt.savefig("../images/Top_10_Countries.png",bbox_inches="tight")
plt.show()
```

```python
df["release_year"].value_counts().sort_index().plot(figsize=(12,5))

plt.title("Netflix Content Growth")
plt.savefig("../images/Netflix_Content_Growth.png",bbox_inches="tight")
plt.show()
```

```python
df["rating"].value_counts().plot(kind="bar")

plt.title("Ratings Distribution")
plt.savefig("../images/Ratings_Distribution.png",bbox_inches="tight")
plt.show()
```

# Insights

- Movies dominate Netflix content.
- USA contributes the largest amount of content.
- Netflix content increased significantly after 2015.
- TV-MA is one of the most common ratings.
- Several columns contain missing values that need cleaning.
